```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ==========================================
# 1. 超参数与环境设置
# ==========================================
BATCH_SIZE = 64
SEQ_LEN = 50        
D_MODEL = 32        
NHEAD = 2           
NUM_LAYERS = 2      
LR = 0.0002
EPOCHS = 600       

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
# 数据生成：正弦波
def get_real_samples(batch_size, seq_len):
    phase = np.random.uniform(0, 2 * np.pi, (batch_size, 1))
    x = np.linspace(0, 4 * np.pi, seq_len)
    data = np.sin(x + phase) 
    return torch.FloatTensor(data).unsqueeze(-1).to(device)
```

```python
# ==========================================
# 2. 网络定义 (标准 Transformer 架构)
# ==========================================
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=100):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]
```

```python
class TransformerGenerator(nn.Module):
    def __init__(self, seq_len, d_model, nhead, num_layers):
        super().__init__()
        self.input_layer = nn.Linear(1, d_model)
        self.pos_encoder = PositionalEncoding(d_model, max_len=seq_len)
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, batch_first=True, activation='relu')
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.output_layer = nn.Linear(d_model, 1)

    def forward(self, z):
        x = self.input_layer(z)
        x = self.pos_encoder(x)
        x = self.transformer(x)
        return self.output_layer(x)
```

```python
class TransformerDiscriminator(nn.Module):
    def __init__(self, seq_len, d_model, nhead, num_layers):
        super().__init__()
        self.input_layer = nn.Linear(1, d_model)
        self.pos_encoder = PositionalEncoding(d_model, max_len=seq_len)
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, batch_first=True, activation='relu')
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(seq_len * d_model, 1),
            nn.Sigmoid() 
        )

    def forward(self, x):
        features = self.input_layer(x)
        features = self.pos_encoder(features)
        features = self.transformer(features)
        return self.classifier(features)
```

```python
# 初始化
netG = TransformerGenerator(SEQ_LEN, D_MODEL, NHEAD, NUM_LAYERS).to(device)
netD = TransformerDiscriminator(SEQ_LEN, D_MODEL, NHEAD, NUM_LAYERS).to(device)
```

```python
criterion = nn.BCELoss()
optimizerG = optim.Adam(netG.parameters(), lr=LR, betas=(0.5, 0.999))
optimizerD = optim.Adam(netD.parameters(), lr=LR, betas=(0.5, 0.999))
```

```python
# ==========================================
# 3. 基础 GAN 预训练循环
# ==========================================
print("第一步：正在并行训练 Transformer GAN...")
for epoch in range(EPOCHS):
    # 训练 D
    netD.zero_grad()
    real_data = get_real_samples(BATCH_SIZE, SEQ_LEN)
    output_real = netD(real_data)
    lossD_real = criterion(output_real, torch.ones(BATCH_SIZE, 1).to(device))
    
    noise = torch.randn(BATCH_SIZE, SEQ_LEN, 1).to(device)
    fake_data = netG(noise)
    output_fake = netD(fake_data.detach())
    lossD_fake = criterion(output_fake, torch.zeros(BATCH_SIZE, 1).to(device))
    
    lossD = lossD_real + lossD_fake
    lossD.backward()
    optimizerD.step()
    
    # 训练 G
    netG.zero_grad()
    output_g = netD(fake_data)
    lossG = criterion(output_g, torch.ones(BATCH_SIZE, 1).to(device))
    lossG.backward()
    optimizerG.step()

    # 打印当前 epoch 的训练信息
    print(f"Epoch [{epoch+1}/{EPOCHS}] "
          f"Loss_D: {lossD.item():.4f} (real: {lossD_real.item():.4f}, fake: {lossD_fake.item():.4f}) "
          f"Loss_G: {lossG.item():.4f}")

print("GAN 基础训练完成！判别器现在已经具备高度的真实/欺骗代码（波形）的分辨能力。")
```

```python
def test_adversarial_attack(netD, netG):
    netD.eval()
    netG.eval()
    
    # 1. 先让生成器生成一段常规的假波形
    noise = torch.randn(1, SEQ_LEN, 1).to(device)
    normal_fake_code = netG(noise).detach()
    
    initial_judgment = netD(normal_fake_code).item()
    print(f"--> [正常混淆代码] 扔给 AI 判别器，AI 识别出它是假数据的置信度为: {(1 - initial_judgment)*100:.2f}%")
    
    # 分离出前 45 个常规 Token (保持不变)
    base_code = normal_fake_code[:, :-5, :].detach()
    
    # 2. 独立初始化 5 个对抗防御 Token，并显式开启梯度
    # 类比看雪文章中 GCG 算法针对攻击后缀进行更新
    poc_tokens = normal_fake_code[:, -5:, :].clone().detach().requires_grad_(True)
    
    # 简单的步长 (学习率)
    step_size = 0.5
    
    # 迭代优化对抗后缀
    for step in range(20):
        # 核心：将基础代码和【带梯度的对抗后缀】动态拼接在一起，重建计算图
        protected_code = torch.cat([base_code, poc_tokens], dim=1)
        
        # 前向传播
        out = netD(protected_code)
        loss = criterion(out, torch.ones(1, 1).to(device))
        
        # 清空残留梯度
        if poc_tokens.grad is not None:
            poc_tokens.grad.zero_()
            
        # 反向传播 (此时梯度链完全畅通)
        loss.backward()
        
        # 沿着梯度负方向更新后缀 (使用 .data 避免破坏计算图跟踪)
        if poc_tokens.grad is not None:
            poc_tokens.data -= step_size * poc_tokens.grad.data
            
    # 3. 再次将附加了 PoC 后缀的代码提交给 AI 判别器
    with torch.no_grad():
        final_protected_code = torch.cat([base_code, poc_tokens], dim=1)
        final_judgment = netD(final_protected_code).item()
        
    print(f"--> [添加对抗PoC后缀后] 再次扔给 AI 判别器，AI 识别置信度变为: {(1 - final_judgment)*100:.2f}%")
    print(f"--> 此时判别器的输出倾向被彻底逆转（成功诱导 AI 拒答或误判）。")
```

```python
test_adversarial_attack(netD, netG)
```
